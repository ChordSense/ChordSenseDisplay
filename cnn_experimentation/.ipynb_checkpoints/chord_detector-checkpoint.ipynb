Dataset Class. Label is the directory where the wav file exists

```python
class ChordDataset(Dataset):
    def __init__(self, chord_root_dir, sr=22050, n_mfcc=13, max_len=238):
        self.sr = sr
        self.n_mfcc = n_mfcc
        self.max_len = max_len
        self.file_paths = []  # was chord_files, but you reference file_paths later
        self.labels = []

        for folder in os.listdir(chord_root_dir):
            folder_path = os.path.join(chord_root_dir, folder)  # was root_dir
            if os.path.isdir(folder_path):
                for file in os.listdir(folder_path):
                    if file.endswith('.wav'):
                        self.file_paths.append(os.path.join(folder_path, file))
                        label = file.split('_')[0]
                        self.labels.append(label)

        # Map string labels to integers
        unique_labels = sorted(set(self.labels))
        self.label_to_index = {label: i for i, label in enumerate(unique_labels)}
        self.num_classes = len(unique_labels)
        self.labels = [self.label_to_index[l] for l in self.labels]

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, idx):
        signal, _ = librosa.load(self.file_paths[idx], sr=self.sr)
        mfcc = librosa.feature.mfcc(y=signal, sr=self.sr, n_mfcc=self.n_mfcc)

        # Pad or truncate to fixed length
        if mfcc.shape[1] < self.max_len:
            pad_width = self.max_len - mfcc.shape[1]
            mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode='constant')
        else:
            mfcc = mfcc[:, :self.max_len]

        mfcc_tensor = torch.tensor(mfcc, dtype=torch.float32).unsqueeze(0)
        mfcc_label = torch.tensor(self.labels[idx], dtype=torch.long)
        return mfcc_tensor, mfcc_label
```

```python
train_path = "Datasets/GuitarChordsV3"
train_data = ChordDataset(train_path)
```

```python
train_size = int(0.8 * len(train_data))
test_size = len(train_data) - train_size
train_set, test_set = random_split(train_data, [train_size, test_size])

print(f"Total samples: {len(train_data)}")
print(f"Train: {train_size}, Test: {test_size}")
print(f"Classes: {train_data.label_to_index}")
```

```python
train_dataloader = DataLoader(train_set, batch_size=16, shuffle=True)
test_dataloader = DataLoader(test_set, batch_size=16, shuffle=False)
```

## Convolutional Neural Network Definition

```python
import torch.nn as nn
import torch.nn.functional as F


class ChordCNN(nn.Module):
    def __init__(self, num_classes=8):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 16, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(16, 32, kernel_size=3, padding=1)
        
        self.pool = nn.MaxPool2d(2)
        
        # After conv1 + pool: (16, 6, 119)
        # After conv2 + pool: (32, 3, 59)
        # Flattened: 32 * 3 * 59 = 5664
        self.fc1 = nn.Linear(32 * 3 * 59, 64)
        self.fc2 = nn.Linear(64, num_classes)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.flatten(1)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
net = ChordCNN(8)
```

### Loss Function and Optimizer

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(net.parameters(), lr=0.001)
```

## Training and Testing Functions

```python
def train(model, dataloader, criterion, optimizer, device):
    model.train()  # set model to training mode (enables gradient tracking)
    running_loss = 0.0
    correct = 0
    total = 0

    for mfccs, labels in dataloader:
        mfccs = mfccs.to(device)
        labels = labels.to(device)

        # Clear gradients from previous batch
        optimizer.zero_grad()
        # Forward pass: get predictions
        outputs = model(mfccs)
        # Compute loss
        loss = criterion(outputs, labels)
        # Backward pass: compute gradients
        loss.backward()
        # Update weights
        optimizer.step()
        # Track metrics
        running_loss += loss.item()
        _, predicted = outputs.max(1)  # get the class with highest score
        total += labels.size(0)
        correct += predicted.eq(labels).sum().item()

    avg_loss = running_loss / len(dataloader)
    accuracy = 100.0 * correct / total
    return avg_loss, accuracy

def test(model, dataloader, criterion, device):
    model.eval()  # set model to evaluation mode (disables dropout etc.)
    running_loss = 0.0
    correct = 0
    total = 0
    with torch.no_grad():  # disable gradient computation for efficiency
        for mfccs, labels in dataloader:
            mfccs = mfccs.to(device)
            labels = labels.to(device)

            outputs = model(mfccs)
            loss = criterion(outputs, labels)

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    avg_loss = running_loss / len(dataloader)
    accuracy = 100.0 * correct / total
    return avg_loss, accuracy
```

## Train model over several epochs

```python
model = ChordCNN(8)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

num_epochs = 50

for epoch in range(num_epochs):
    print(f"Starting epoch number {epoch}")
    train_loss, train_acc = train(model, train_dataloader, criterion, optimizer, device)
    test_loss, test_acc = test(model, test_dataloader, criterion, device)

    print(f"Epoch {epoch+1}/{num_epochs} | "
          f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.1f}% | "
          f"Test Loss: {test_loss:.4f}, Test Acc: {test_acc:.1f}%")
```

## Test on my recordings

```python
def predict_chord(model, file_path, label, label_to_index, sr=22050, n_mfcc=13, max_len=238):
    index_to_label = {i: l for l, i in label_to_index.items()}

    signal, _ = librosa.load(file_path, sr=sr)
    mfcc = librosa.feature.mfcc(y=signal, sr=sr, n_mfcc=n_mfcc)

    if mfcc.shape[1] < max_len:
        pad_width = max_len - mfcc.shape[1]
        mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode='constant')
    else:
        mfcc = mfcc[:, :max_len]

    mfcc_tensor = torch.tensor(mfcc, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
    mfcc_tensor = mfcc_tensor.to(device)

    model.eval()
    with torch.no_grad():
        output = model(mfcc_tensor)
        probabilities = torch.softmax(output, dim=1)
        confidence, predicted = probabilities.max(1)

    predicted_chord = index_to_label[predicted.item()]
    conf = confidence.item() * 100
    correct = predicted_chord == label

    print(f"Actual: {label}")
    print(f"Predicted: {predicted_chord} ({conf:.1f}% confidence)")
    print(f"Correct: {correct}")
    return predicted_chord, conf, correct
```

```python
predict_chord(model, "Datasets/MyRecordings/Recording1.wav", "G", train_data.label_to_index)
```
